```python
# Imports
import os
import yake

# Instantier l'extracteur de mots clés
kw_extractor = yake.KeywordExtractor(lan="fr", top=50)
kw_extractor

# Lister les Fichiers
data_path = "C:/Users/Admin/Tac_Test/tac/data/txt/" 
files = [f for f in os.listdir(data_path) if f.endswith('.txt')]

# Choisir et lister les fichiers d'une année
year = 1889
files_1889=[f for f in files if str(year) in f]

#Choisir un fichier

fichier = files_1889[0]  # Premier fichier de l'année 1889
text = open(os.path.join(data_path, fichier), 'r', encoding='utf-8').read()
text[:500]  # Affiche les 500 premiers caractères

# Extraire les mots clés
keywords = kw_extractor.extract_keywords(text)
keywords

```

```python
#Ne garder que les bigrammes pour l'année 1889
for f in sorted(files_1889):
    text = open(os.path.join(data_path, f), 'r', encoding="utf-8").read()
    keywords = kw_extractor.extract_keywords(text)
    kept = []
    for kw, score in keywords:
        words = kw.split()
        if len(words) == 2:
            kept.append(kw)
    print(f"{f} mentions these keywords: {', '.join(kept)}...")
```
