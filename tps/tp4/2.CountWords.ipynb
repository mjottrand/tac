## Compter le nombre de mots dans l'ensemble du corpus

le fichier 8 est lus précis pour le comptage de mot (tokens)...

```python
file_path = '../../data/all.txt'

with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
    text = f.read()

num_lines = text.count('\n') + 1
num_words = len(text.split())
num_chars = len(text)

print(f"Lignes : {num_lines}")
print(f"Mots   : {num_words}")
print(f"Caractères : {num_chars}")
```
